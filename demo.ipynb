```python
#importation des boilerplate
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#ajout des données en montrant le chemin
files= [file for file in os.listdir (r'\Users\Caplogy_1010\Desktop\DATASET')]
for file in files:
    print (file)
```

```python
path=r'\Users\Caplogy_1010\Desktop\DATASET'
#initier une base de donné et ensuite afficher toutes les donnnees
unseulefichiercsv_data=pd.DataFrame()

for file in files:
    Current_data=pd.read_csv(path+'/'+file)
    unseulefichiercsv_data=pd.concat([unseulefichiercsv_data,Current_data])
    print(unseulefichiercsv_data)
```

```python
#génerer le fichier csv contenant tous les fichiers 
unseulefichiercsv_data.to_csv(path+'/unseulefichiercsv_data.csv', index=False)
```

```python
#afficher les types dans le fichier csv contenant tous les données
unseulefichiercsv_data.dtypes
```

```python
#lors de l'examen des données, concernent l’élimination des valeurs indéfinies
unseulefichiercsv_data.isnull().sum()
```

```python
# les données vide cleanning
unseulefichiercsv_data=unseulefichiercsv_data.dropna(how='all')
unseulefichiercsv_data.shape
```

```python
unseulefichiercsv_data
```

```python
unseulefichiercsv_data.drop_duplicates()
```

```python
#etape 2 
def month(x):
    return str(x).split('/')[0]
```

```python
unseulefichiercsv_data['Month']=unseulefichiercsv_data['Order Date'].apply(month)
```

```python
unseulefichiercsv_data['Month'].unique()
```

```python
unseulefichiercsv_data=unseulefichiercsv_data[unseulefichiercsv_data['Month']!='Order Date']
unseulefichiercsv_data=unseulefichiercsv_data[unseulefichiercsv_data['Month']!='nan']

unseulefichiercsv_data['Month'].unique()
```

```python
#changement des types 
unseulefichiercsv_data["Month"]=unseulefichiercsv_data['Month'].astype(int)
unseulefichiercsv_data.dtypes
```

```python
unseulefichiercsv_data['Price Each']=unseulefichiercsv_data['Price Each'].astype(float)
unseulefichiercsv_data['Quantity Ordered']=unseulefichiercsv_data['Quantity Ordered'].astype(int)
unseulefichiercsv_data['Order Date']=pd.to_datetime(unseulefichiercsv_data['Order Date'])
unseulefichiercsv_data.dtypes
```

```python
unseulefichiercsv_data.rename(columns={'Purchase Address' :'Adress'}, inplace=True)
```

```python
unseulefichiercsv_data
```

```python
#•	Requête 1: Quel a été le meilleur mois de vente
#•	Commencez par rajouter une colonne Total dont les valeurs sont calculées comme suit : ‘’ Quantity Ordered’’*’’ Price Each’’
unseulefichiercsv_data['Total']=unseulefichiercsv_data['Quantity Ordered']*unseulefichiercsv_data['Price Each']
```

```python
months=range(1,13)
plt.bar(months,unseulefichiercsv_data.groupby('Month')['Total'].sum())
plt.xticks(months)
plt.ylabel('Total')
plt.xlabel('Month')
plt.show()
#le mois avec le plus de vente est le mois 12
```

```python
unseulefichiercsv_data.groupby('Month')['Total'].sum()
```

```python
#•	Requête 2 : quelle est la cité qui a enregistré le maximum de revenus ?
unseulefichiercsv_data
```

```python
def city(x):
    return x.split(',')[1]
unseulefichiercsv_data['city']=unseulefichiercsv_data['Adress'].apply(city)
unseulefichiercsv_data.groupby('city')['city'].count()
```

```python
plt.bar(unseulefichiercsv_data.groupby('city')['city'].count().index,unseulefichiercsv_data.groupby('city')['city'].count())
plt.xticks(rotation='vertical')
plt.ylabel('commandes recus')
plt.xlabel('city nom')
plt.show()
#la cite de san francisco
```

```python
#•	Requête 3 : À quelle heure devrions-nous afficher des publicités pour maximiser la probabilité d'achat du produit par le client?
unseulefichiercsv_data['Heure']=pd.to_datetime(unseulefichiercsv_data['Order Date']).dt.hour
keys=[]
hours=[]
for key,hour in unseulefichiercsv_data.groupby('Heure'):
    keys.append(key)
    hours.append(len(hour))
hours    
plt.grid()
plt.plot(keys,hours)
plt.xlabel('heure de la journée')
plt.ylabel('nombre des commandes')
```

```python
#•	Requête 4 : Quel est le produit le plus vendu 
unseulefichiercsv_data.groupby('Product')['Quantity Ordered'].sum()
unseulefichiercsv_data.groupby('Product')['Quantity Ordered'].sum().plot(kind='bar')
```
